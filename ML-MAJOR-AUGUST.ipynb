```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df['income']= label_encoder.fit_transform(df['income'])
  
df['income'].unique()
```

```python
df
```

```python
x=df.drop('income',axis=1)
y=df['income']
```

```python
x_encoded=pd.get_dummies(x,columns=['Workclass','occupation'],drop_first=True)
x_encoded
```

```python
y
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True,cmap='Accent')
```

# Decision Tree

```python
from sklearn.model_selection import train_test_split

```

```python
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y,test_size=0.25)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier( criterion='gini',random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
c=model.score(x_test,y_test)
c
```

```python
model.score(x_train,y_train)
```

```python
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score,accuracy_score,precision_score,recall_score,f1_score
cm_train=confusion_matrix(y_train,y_pred_train)
cm_train
```

```python
b=classification_report(y_train,y_pred_train)
print(b)
```

```python
cm_test=confusion_matrix(y_test,y_pred_test)
cm_test
```

```python
b=classification_report(y_test,y_pred_test)
print(b)
```

```python
misclassification_rate=1-c
misclassification_rate
```
